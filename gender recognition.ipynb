```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(dataset,label,test_size=0.20)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
ytrain=to_categorical(ytrain,num_classes=2)
```

```python
ytrain
```

```python
ytest=to_categorical(ytest,num_classes=2)
```

```python
ytest
```

```python
from tensorflow.keras import Sequential
```

```python
model=Sequential()
```

```python
model.add(layers.Input(shape=(size,size,3)))
```

# first conv2D layer

```python
model.add(layers.Conv2D(32,(3,3),activation="relu",padding="same"))
model.add(layers.MaxPooling2D(pool_size=(3,3)))
model.add(layers.BatchNormalization(axis=-1))
model.add(layers.Dropout(0.25))
```

# second conv2D layer

```python
model.add(layers.Conv2D(64,(3,3),activation="relu",padding="same"))
model.add(layers.MaxPooling2D(pool_size=(3,3)))
model.add(layers.BatchNormalization(axis=-1))
model.add(layers.Dropout(0.25))
```

# third conv2d layer

```python
model.add(layers.Conv2D(64,(3,3),activation="relu",padding="same"))
model.add(layers.MaxPooling2D(pool_size=(3,3)))
model.add(layers.BatchNormalization(axis=-1))
model.add(layers.Dropout(0.25))
```

# fourth conv2D layer

```python
model.add(layers.Conv2D(128,(3,3),activation="relu",padding="same"))
model.add(layers.MaxPooling2D(pool_size=(3,3)))
model.add(layers.BatchNormalization(axis=-1))
model.add(layers.Dropout(0.25))
```

# Dense layer

```python
model.add(layers.Flatten())
model.add(layers.Dense(1024,activation="relu"))
model.add(layers.BatchNormalization(axis=-1))
model.add(layers.Dropout(0.25))
```

```python
model.add(layers.Dense(2,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model
```

```python
model.fit(Xtrain,ytrain,epochs=25,verbose=1)
```

```python
model.evaluate(Xtest,ytest)
```
